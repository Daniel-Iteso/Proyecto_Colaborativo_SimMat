```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-10, 10, 0.1)
y = x ** 3
plt.plot(x, y, label='$y=x^3$')
plt.legend()
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.title('Ilustración función $y=x^3$')
plt.show()
```
